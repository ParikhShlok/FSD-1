```python
class complex:#for operator overloading(arithmetic operator)
    def __init__(self,r,i):
        self.real=r
        self.imagenary=i
    def __add__(self,other):
        real=self.real+other.real
        imagenary=self.imagenary+other.imagenary
        return f"{real}+{imagenary}i"
c1=complex(1,4)
c2=complex(3,5)
print(c1+c2)
```

```python
class A:
    def __init__(self,a):
        self.a=a
    def __gt__(self,other):
        if self.a>other.a:
            return f"{self.a} is greater"
        else:
            return f"{other.a} is greater"
s=A(5)
s1=A(3)
print(s1)
```

### Method Overloading

Python does not support method overloading by default. If you define multiple methods with the same name, only the latest definition will be used.
python does not support traditional method overloading

```python
class addition:
    def sum(self,a,b):
        return a+b
    def sum(self,a,b,c=2):#default argument
        return a+b+c
o=addition()
print(o.sum(2,3))
print(o.sum(12,35,21))
```

```python
class addition:
    def sum(self,a,b):
        return a+b
    def sum(self,*args):#using argument
        sum=0
        for i in args:
            sum=sum+i
        return sum
o=addition()
print(o.sum(2,3))
print(o.sum(12,35,21))
```

### Method Overriding(inheritance)

#### SINGLE INHERITANCE

```python
class parent:
    def sleep(self):
        print("sleep at 9")
class child(parent):
    def sleep(self):
        print("sleep at 2")
    def eat(self):
        print("eat at 4")
obj=parent()
obj.sleep()
```

```python
class parent:
    def sleep(self):
        print("sleep at 9")
class child(parent):
    def sleep(self):
        print("sleep at 2")
        super().sleep()
    def eat(self):
        print("eat at 4")
obj=child()
obj.sleep()
obj.eat()
```

```python
class parent:
    def sleep(self):
        print("sleep at 9")
class child(parent):
    def wake(self):
        print("wake at 2")
    def eat(self):
        print("eat at 4")
obj=child()
obj.sleep()
obj.eat()
obj.wake()
print(child.mro())
print(parent.mro())
```

#### MULTIPLE INHERITANCE

```python
#MULTIPLE PARENT CLASS AND SINGLE CHILD CLASS
class father:
    def sleep(self):
        print("sleep at 9")
    def swim(self):
        print("swim at 6")
class mother:
    def cook(self):
        print("cook good")
    def sleep(self):
        print("sleep at 10")
class child(father,mother):#j class pela lakhyo hoi tema jase
    def run(self):
        print("run very fast")
obj=child()
obj.sleep()
print(child.mro())
```

#### MULTILEVEL INHERITANCE

```python
class grandfather:
    def sleep(self):
        print("sleep at 9")
class father(grandfather):
    def swim(self):
        print("swim at 6")
class son(father):
    def eat(self):
        print("eat at 4")
obj=son()
obj.sleep()
obj.swim()
obj.eat()
print(son.mro())
```

#### HIERARCHICAL INHERITANCE

```python
class parents:
    def display(self):
        print("display from parent")
class child(parents):
    def fun(self):
        print("fun from child class")
class child2(parents):
    def show(self):
        print("show from child2 class")
        
obj=child()
obj.display()
obj.fun()
print(child.mro())
obj2=child2()
obj2.display()
obj2.show()
print(child.mro())
```

```python
class A:
    def display(self):
        print("multiple inheritance class 1")
class B:
    def fun(self):
        print("multiple inheritance class 2")
class C(A,B):
    def show(self):
        print("mulitple inheritance child")
class D(C):
    def show1(self):
        print("hierarchical inheritance child 1")
class E(C):
    def show2(self):
        print("hierarchical inheritance child 2")
obj=E()
obj.display()
print(D.mro())
print(E.mro())
```
