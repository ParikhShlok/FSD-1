```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,2,4,5,7,8])
y=np.array([3,7,6,12,9,3])
x1=([0,1,2,3,5,7,9])
y1=([12,7,12,3,3,5,7])
plt.subplot(1,2,1)#row,column,grapgh number
plt.plot(x,y,marker="o",color='r',ms=10,mec='g',mfc='m',ls='-.',lw=5)
plt.subplot(1,2,2)
plt.plot(x1,y1,marker='*',color='b',ms=15,mec='r',mfc='y',ls=':',lw=3)
plt.show()
```

## Area plot & Stacked Area plot

```python
import matplotlib.pyplot as plt
import numpy as np
x=range(1,6)#x=np.array([1,2,3,4,5])
y=[1,4,7,7,4]
plt.fill_between(x,y,color="r",alpha=0.5)
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=([2,4,5,6,8])
y=([4,6,12,8,12])
plt.plot(x,y)
y1=([8,2,2,14,6])
plt.plot(x,y1)
plt.fill_between(x,y,color="coral")
plt.fill_between(x,y1,color="teal")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=range(1,13)
y=([5,9,6,6,10,7,6,4,4,5,6,4])
plt.plot(x,y,c="g")
y1=([6,6,8,3,6,9,7,8,6,6,4,8])
plt.plot(x,y1,c='r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=range(1,13)
y=np.array([5,9,6,6,10,7,6,4,4,5,6,4])
plt.plot(x,y,c="g")
y1=np.array([6,6,8,3,6,9,7,8,6,6,4,8])
plt.plot(x,y1,c='r')
plt.fill_between(x,y,y1,where=(y>y1),color='g',alpha=0.5,label='positive')
plt.fill_between(x,y,y1,where=(y<y1),color='r',alpha=0.5,label='negative')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=range(1,13)
y=np.array([5,9,6,6,10,7,6,4,4,5,6,4])
plt.plot(x,y,c="g")
y1=np.array([6,6,8,3,6,9,7,8,6,6,4,8])
plt.plot(x,y1,c='r')
plt.fill_between(x,y,y1,where=(y>y1),color='g',alpha=0.5,label='positive',interpolate=True)
plt.fill_between(x,y,y1,where=(y<y1),color='r',alpha=0.5,label='negative',interpolate=True)
plt.legend()
plt.show()
```

## Stack area plot

```python
import matplotlib.pyplot as plt
import numpy as np
x=range(1,6)
y=np.array([2,3,3,1,2])
y1=np.array([2,2,3,2,4])
y2=np.array([2,1,3,2,1])
plt.stackplot(x,y,y1,y2,labels=['apple','banana','cherry'],colors=['coral','teal','royalblue'])
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=range(1,6)
y=np.array([2,2,2,2,2])
y1=np.array([2,2,2,2,2])
y2=np.array([2,2,2,2,2])
plt.stackplot(x,y,y1,y2,colors=['yellow','red','black'])
plt.title('germany')
plt.show()
```
