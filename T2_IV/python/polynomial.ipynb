# Single/Simple Regression

````python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
````

````python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
x=pd.DataFrame(x)
y=pd.Series(y)
````

````python
print(x.shape)
print(y.shape)
````

````python
poly=PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x)
````

````python
print(x_poly)
````

````python
lr=LinearRegression()
lr.fit(x_poly,y)
````

````python
ypred=lr.predict(x_poly)
ypred
````

````python
print(r2_score(y,ypred))
````

````python
print(lr.coef_)
````

````python
print(lr.intercept_)
````
