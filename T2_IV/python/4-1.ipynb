````python
df=pd.read_csv("placement_03-06.csv")
df
````

````python
df.info()
````

````python
df.isna().sum()
````

````python
x=df[['cgpa']]#x always in data frame
x
````

````python
y=df['package']#y is always in series
y
````

````python
x.shape#(row,column)
````

````python
y.shape
````

````python
from sklearn.model_selection import train_test_split
````

```
sequence of initialize train and test is fixed like in that pair first x 
train size=0.8
test size=0.2
```

````python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=1)
````

````python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
````

````python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)#model ready
````

````python
#testing
ypred=lr.predict(x_test)
print(ypred)
````

````python
#make dataframe
diff=pd.DataFrame({"Actual":y_test,"Prediction":ypred})
diff
````

````python
print("CO-EFICIENT:",lr.coef_)#VALUE OF COEFICIENT
````

````python
print("INTERCEPT:",lr.intercept_) #VALUE OF INTERCEPT
````

````python
from sklearn import metrics
print("MAE:",metrics.mean_absolute_error(y_test,ypred))
print("MSE:",metrics.mean_squared_error(y_test,ypred))
print("R2_Score:",metrics.r2_score(y_test,ypred))
````
