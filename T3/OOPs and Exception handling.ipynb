```python
#static method
class student:
    def __init__(self,name,rollno):
        self.name=name
        self.rollno=rollno
    @staticmethod #decorator
    def abc():
        print("hello")
a=student("tejas",10)
print(a.name)
a.abc()
```

```python
class atm():
    def __init___(self):
        self.pin=""
        self.balance=0
        self.wel()
    def wel(self):
        choose=input('''1.create a pin 
        2. new pin 
        3.check balance 
        4.withdraw
        5.exit''')
        if choose=="1":
            self.create()
        elif choose=="2":
            self.newpin()
        elif choose=="3":
            self.check()
        elif choose=="4":
            self.withdraw()
        else:
            print("exit")
    def create(self):
        pin=input("enter pin")
        self.pin=pin
        bal=input("enter amount")
        self.balance=bal
        print("pin created succefully")
        self.wel()
    def newpin(self):
        a=input("enter old pin")
        if a==pin:
            self.pin=a
        else:
            print('enter correct pin')
        self.wel()
    def check(self):
        b=input("enter pin")
        if self.pin==b:
            print(self.balance)
        self.wel()
    def withdraw(self):
        c=input("enter pin")
        if c==self.pin:
            d=int(input("enter withdraw amount"))
            if d<=self.balance:
                self.balance=self.balance-d
                print(self.balance)
            else:
                print("enter valid amount")
        self.wel()
s=atm()
```
