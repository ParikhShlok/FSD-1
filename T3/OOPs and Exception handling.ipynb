```python
class Employee:
    def __init__(self,name,age,salary):
        self.name=name
        self.age=age
        self.salary=salary
    def emp(self):
        print(f"employee name is {self.name} age is {self.age} and salary {self.salary} ")
keval=Employee("manish",54,5325530)
# keval=Employee("hgds",54)
keval=Employee(255,"Kdkj",6853563)
keval.emp()
```

```python
class Employee:
    def __init__(self,name,age,salary,increment):
        self.name=name
        self.age=age
        self.salary=salary
        self.increment=increment
    def emp(self):
        self.salary=self.salary*self.increment
        print(f"salary is {self.salary}")
keval=Employee("manish",54,5325530,1.5) 
keval.emp()
hiren=Employee("manish",54,5325530,1.7) 
hiren.emp()
```
