```python
#WAP to copy odd lines of one file into another file name as odd.txt
f=open("hello.txt","r")
a=f.readlines()
f1=open("odd.txt",'w+')
for i in range(len(a)):
    if i%2==0:
        f1.write(str(a))
f.close()
f1.close()
```

```python
f=open("compare1.txt","r")
f1=open("compare2.txt","r")
a=f.read()
b=f1.read()
line,pos=1,1
for i in range(len(a)):
    if a[i]==b[i]:
        pos+=1
        if a[i]=="\n":
            line+=1
            pos=1
    else:
        print(f"line: {line}, pos: {pos}")
        pos+=1
f.close()
f1.close()
```

```python
#WAP to display all line of a file axcept those that have # sign
#aama avu karvanu chhe k # pachi no data biji file ma javo jove and te line print no thavi joi a
f=open("i1.txt","r")
f1=open("o2.txt","w+")
while True:
    a=f.readline()
    if len(a)!=0:
        if a[0]=='#':
            continue
        else:
            if '#' in a:
                for i in range(1,len(a)):
                    if a[i]=='#':
                        f1.write(a[0:i]+"\n")
            else:
                f1.write(a)
    else:
        break
f.close()
f1.close()
```

```python
#WAP to count words, character and space from given text file
f=open("count.txt",'r')
a=f.read()
b=a.split()
char=len(a)-a.count(" ")-a.count("\n")
space=a.count(" ")
print(f"words={len(b)}")
print("character=",char)
print("space=",space)
```
