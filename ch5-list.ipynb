### ALIASING AND CLONNING OF LIST

```python
#ALIASING
l=[1,2,3,4]
temp=l
temp[1]=100
print(temp)
print(l)
```

```python
#CLONNIG trick:-1(create a clonning by using slice operator)
l=[1,2,3,4]
temp=l[:]
temp[1]=100
print(l)
print(temp)
```

```python
#CLONNIG trick:-2(create a clonning by using copy())
l=[1,2,3,4]
temp=l.copy()
temp[1]=100
print(l)
print(temp)
```

```python
#list comprehension
l=[i**2 for i in range(0,10)]
print(l)
```

```python
#tuple
t=tuple(i**2 for i in range(0,10))
print(t)
```

```python
fruit=["apple","kivy","banana","ananas","blue berry"]
for i in fruit:
    if 'a' not in i:
        fruit.remove(i)
print(fruit)
```

```python
fruit=["apple","kivy","banana","ananas","blue berry"]
l=[]
for i in fruit:
    if 'a' in i:
        l.append(i)
print(l)
```

```python
#comrehension code
fruit=["apple","kivy","banana","ananas","blue berry"]
l=[i for i in fruit if 'a' in i]
print(l)
```

```python
l=[x**2 for x in range(0,10)]
even=[i for i in l if i%2==0]
print(even)
```
