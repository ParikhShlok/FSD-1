```python
#copy()
d={10: 'abc', 6: 'baba', 9: 'bacbc', 20: 'manish'}
d1=d.copy()
d1[9]='popat'
print(d)
print(d1)
```

```python
#update()
d={10: 'abc', 6: 'baba', 9: 'bacbc', 20: 'manish'}
d1={39: 'hey', 4: 'good', 9: 'popat', 2: 'manish'}
d.update(d1)
print(d)
```
