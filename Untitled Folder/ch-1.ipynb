```python
#for removing empty data
#by filling empty data using null(0)
#by filling empty data using mean, median
```

```python
import pandas as pd
import numpy as np
data={
    "name":["ravi",'kishor','tej',np.nan,np.nan,'vikki'],
    "salary":[20000,30900,np.nan,np.nan,62000,np.nan],
    "except":[np.nan,10900,30000,np.nan,np.nan,40000],
    'region':['bopal','shyaamal',np.nan,np.nan,'maninagar','vasna']
}
dataset=pd.DataFrame(data)
dataset
```

```python
dataset.isna()
```

```python
dataset.isna().sum()
```

```python
dataset.dropna()
```

```python
dataset.dropna(axis=0,how='all',thresh=3)
```

```python
dataset.dropna(subset=['salary'])
```

```python
dataset.dropna(subset=['salary','except'])
```

```python
dataset.dropna(subset=['salary','name','region'])
```

```python
dataset
```

```python
dataset.dropna(inplace=True)
```

```python
dataset
```

```python
data={
    "name":["ravi",'kishor','tej',np.nan,np.nan,'vikki'],
    "salary":[20000,30900,np.nan,np.nan,62000,np.nan],
    "except":[np.nan,10900,30000,np.nan,np.nan,40000],
    'region':['bopal','shyaamal',np.nan,np.nan,'maninagar','vasna']
}
dataset=pd.DataFrame(data)
dataset
```

```python
dataset.fillna(0)
```

```python
dataset.describe()
```

### fill the null value by using mean and median value

```python
dataset['salary'].fillna(dataset['salary'].mean())
```

```python
dataset['salary'].fillna(dataset['salary'].median())
```
