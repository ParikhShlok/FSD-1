```python
import pandas as pd
import numpy as np
```

```python
df1=pd.DataFrame({'A':[1,2],'B':[3,4]})
df2=pd.DataFrame({'A':[5,6],'B':[7,8]})
print(df1)
print("--------------")
print(df2)
```

```python
pd.concat([df1,df2])
```

```python
df1=pd.DataFrame({'A':[1,2],'B':[3,4]})
df2=pd.DataFrame({'A':[5,6],'B':[7,8]})
pd.concat([df1,df2],join='inner')
```

```python
df1=pd.DataFrame({'A':[1,2],'B':[3,4]})
df2=pd.DataFrame({'A':[5,6],'B':[7,8]})
pd.concat([df1,df2],join='outer')
```

```python
df1=pd.DataFrame({'A':[1,2],'B':[3,4]})
df2=pd.DataFrame({'A':[5,6],'B':[7,8]})
pd.concat([df1,df2],axis=1)
```

```python
pd.concat([df1,df2],ignore_index=True)
```

```python
df1=pd.DataFrame({'D':[1,2],'B':[3,4]})
df2=pd.DataFrame({'E':[5,6],'A':[7,8]})
pd.concat([df1,df2],sort=True)
```

```python
employees=pd.DataFrame({
    'emp_id':[1,2,3,4],
    'name':['A','B','C','D'],
    'dept_id':[10,20,10,30]
})
print(employees)

department=pd.DataFrame({
    'dept_id':[10,20,40],
    'name':['IT','HR','Finance']
})
print(department)
```

```python
pd.merge(employees,department,on='dept_id',how='outer')
```

```python
pd.merge(employees,department,on='dept_id',how='inner')
```

```python
pd.merge(employees,department,on='dept_id',how='left')
```

```python
pd.merge(employees,department,on='dept_id',how='right')
```

```python
employees=pd.DataFrame({
    'emp_id':[1,2,3,4],
    'name':['A','B','C','D'],
    'dept_aid':[10,20,10,30]
})
print(employees)
```

```python
pd.merge(employees,department,left_on='dept_aid',right_on="dept_id",sort=True,how='outer',indicator=True,suffixes=("_emp","_dept"))
```

```python
pd.merge(employees,department,left_index=True,right_index=True)
```
