# HEATMAP :

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data=np.random.randint(low=1,high=100,size=(10,10))
data
```

```python
sns.heatmap(data,annot=True,vmin=30,vmax=70,cmap='viridis',lw=5,linecolor='black',center=50,xticklabels=False,yticklabels=False)
plt.show()
```

# NetworkX :

```python
import networkx as nx
```

```python
G=nx.Graph()
G.add_node(1)
G.add_nodes_from([2,3])
G.add_nodes_from(range(4,7))
G.add_edge(1,2)
G.add_edges_from([(2,3),(3,4),(5,5),(4,5),(1,6),(4,2),(6,1)])
nx.draw_networkx(G,node_size=500,node_color='red',edge_color='blue',width=5)
plt.show()
```

```python
G=nx.DiGraph()
G.add_node(1)
G.add_nodes_from([2,3])
G.add_nodes_from(range(4,7))
G.add_edge(1,2)
G.add_edges_from([(2,3),(3,4),(5,5),(4,5),(1,6),(4,2),(6,1)])
nx.draw_networkx(G,node_size=500,node_color='red',edge_color='blue',width=5)
plt.show()
```

# NetworkX diagram using Adjacency Matrix :

```python
G=nx.DiGraph()
l=[[0,1,0,1],[0,0,0,1],[1,0,0,0],[1,0,1,0]]
G.add_nodes_from(range(4))
for i in range(4):
    for j in range(4):
        if l[i][j]==1:
            G.add_edge(i,j)
nx.draw_networkx(G,node_color='green',node_size=500,node_shape='*',edge_color='m',width=5)
plt.show()
```

```python
G=nx.DiGraph()
l=[[1,0,0,1,0,1,1],[0,1,1,1,1,1,0],[1,1,1,0,1,0,1]]
G.add_nodes_from(range(3))
for i in range(3):
    for j in range(7):
        if l[i][j]==1:
            G.add_edge(i,j)
nx.draw_networkx(G,node_color='b',node_size=500,node_shape='h',edge_color='g',width=3)
plt.show()
```

# SCATTER PLOT :

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('auto-mpg.csv')
df['horsepower']=pd.to_numeric(df['horsepower'],errors='coerce')
df=df.dropna(subset=['horsepower'])
sns.scatterplot(data=df,x='horsepower',y='mpg',hue='origin',palette='viridis')
plt.title('MPG vs Hporsepower')
plt.show()
```
