```python
data={
    'width':[1,2],
    'height':[3,4]
    }
df=pd.DataFrame(data)
df
```

```python
def get_stat(x):
    return [x['width']*x['height'],2*(x['width']+x['height'])]
df
```

```python
df.apply(get_stat,axis=1,result_type='expand')
```

```python
df.apply(get_stat,axis=1,result_type='broadcast')
```

```python
df.apply(get_stat,axis=1,result_type='reduce')
```

```python
df.apply(get_stat,axis=1,result_type=None)
```
